### Adress Type 

```python
train.loc[train["Address"].str.contains("/"), "AdressType"] = "Crossroad"
train.loc[train["Address"].str.contains("of"), "AdressType"] = "Block"
train.head(10)
```

```python
train[train["AdressType"].isnull()]
```

```python
test.loc[test["Address"].str.contains("/"), "AdressType"] = "Crossroad"
test.loc[test["Address"].str.contains("of"), "AdressType"] = "Block"
test.head(10)
```

```python
train.loc[train["AdressType"]=="Crossroad", "Address_encode"] =0
train.loc[train["AdressType"]=="Block", "Address_encode"] =1
train.head()
```

```python
test.loc[test["AdressType"]=="Crossroad", "Address_encode"] =0
test.loc[test["AdressType"]=="Block", "Address_encode"] =1
test.head()
```

## Encoding 

- Encode PdDistrict

```python
pd_district_list = train["PdDistrict"].unique()
print(len(pd_district_list))
pd_district_list
```

get_dummies는 꽤나 위험하다

```python
for pd_district in pd_district_list :
    train["PdDistrict-" + pd_district] = train["PdDistrict"]==pd_district
train.head()
```

```python
for pd_district in pd_district_list :
    test["PdDistrict-" + pd_district] = test["PdDistrict"]==pd_district
test.head()
```

```python
pd_district_list = "PdDistrict-" + pd_district_list
pd_district_list
```

## features 선택 

```python
feature_names = ["X", "Y", "Dates-hour", "Dates-month", 
                 "Address_encode"]
feature_names = feature_names + list(pd_district_list)
feature_names
```

```python
X_train = train[feature_names]
print(X_train.head())
X_test = test[feature_names]
print(X_test.head())
```

```python
y_train = train["Category"]
print(y_train.head())
```

## Model 생성

```python
!pip install xgboost
```

```python
from sklearn.model_selection import cross_val_score
import xgboost as xgb

model = xgb.XGBClassifier(objective="mult:softprob",
                         n_estimators = 50,
                         learning_rate = 1.0,
                         max_depth = 5,
                         max_delta_step = 1,
                         nthread = -1,
                         seed = 0)

score = cross_val_score(model, X_train, y_train, scoring="neg_log_loss").mean()

print("Score = {0:5f}".format(score))
```

## 저장 

```python
print(X_train.shape)
print(X_test.shape)
```

```python
model.fit(X_train, y_train)
predictions = model.predict_proba(X_test)
predictions[:5]
```

```python
print(predictions.shape)
print(test.shape)
```

```python
submission = pd.read_csv("../sf-crime/sampleSubmission.csv", index_col=["Id"])

for i, column in enumerate(submission.columns) :
    submission[column] = predictions[:,i]

print(submission.shape)
submission.head()
```

```python
submission.to_csv("sf-crime_submission.csv")
```
