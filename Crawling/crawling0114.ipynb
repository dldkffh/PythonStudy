```python
url = 'https://www.google.com/search?'

#params: 쿼리스트링 작성 / 딕셔너리 형태 작성
res = req.get(url, params={'q':'빅데이터'})
print(res.text)
```

```python
url = 'https://search.naver.com/search.naver?'

#1.웹 페이지 요청
#헤더 설정으로 경고를 안받을 수 있음, 사람인척ㅋㅋ
res = req.get(url, params={'query':'빅데이터'}, 
              headers={'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/71.0.3578.98 Safari/537.36'})

#2. 응답받은 페이지를 제어할 수 있도로 bs를 이용하여 변환
#bs : 응답데이터, 변환프로그램
#변환 프로그램 종류 : lxml > html.parser > html5lib
soup = bs(res.content, 'html.parser')
```

```python
#title, body, meta, 등 원하는 부분 검색 가능
#.text : 텍스트만 가져오기
soup.find('title').text 
```

```python
soup.find('title', id="id값") 
soup.find('title', class_="class값")
```

```python
#div태그 중에서 id가 main_pack 요소를 검색
soup.find('div', id="main_pack")

#div태그 중에서 class가 main_pack, sub_pack, thre_pack을 가진 요소를 검색
soup.find('div', class_="main_pack") #'main_pack sub_pack thre_pack'
```

```python
soup.find('a', class_='lnk_tit').text
```

ul._related_keyword_ul a
ul._related_keyword_ul > li > a

```python
#.select : CSS선택자 -> 반환형태 : 리스트 = findAll(요소명, id or class_)
#select_one : CSS선택자 -> 반환형태 : 요소 한개 = find(요소명, id or class_)
aTags = soup.select('ul._related_keyword_ul > li > a')
print(aTags[0].text)
print(aTags[1].text)
print(aTags[2].text)
print(aTags[3].text)
print(aTags[4].text)
```

```python
url = 'https://news.naver.com/main/read.nhn?mode=LSD&mid=shm&sid1=102&oid=081&aid=0002970126'
res = req.get(url, headers={'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/71.0.3578.98 Safari/537.36'})
soup = bs(res.content, 'html.parser')
```

```python
soup.find('h3', id='articleTitle')
```

```python
soup.select_one('h3#articleTitle') #select와 같음 대신 하나만 가져오는 거
```

```python
title = soup.select_one('h3#articleTitle')
```

```python
#script 태그 검색 후 삭제
scripts = soup.findAll('script')
len(scripts)
```

```python
for script in scripts :
    script.extract()
```
