# 인스타그램 정보 수집
### 해시태그정보 수집
1. 게시물 클릭
2. 현재 게시물 페이지를 bs로 변환
3. #기준으로 수집

```python
keyword = '마카롱'
url = 'https://www.instagram.com/explore/tags/' + keyword
driver = webdriver.Chrome()
driver.get(url)
```

```python
driver.find_element_by_class_name('_9AhH0').click()
```

```python
all_hastage = []
for i in tn(range(10)) :
    soup = bs(driver.page_source, 'html.parser') #수프 만들기
    hastage = soup.select('h2._6lAjh + span a') #해시태그 수집 
    for i in range(len(hastage)) :
        all_hastage.append(hastage[i].text) #리스트에 추가
    driver.find_element_by_class_name('coreSpriteRightPaginationArrow').click() #오른쪽으로 넘기기
    time.sleep(1)
```

```python
len(all_hastage), all_hastage
```

```python
hastage_list = []
for i in tn(range(10)) :
    soup = bs(driver.page_source, 'html.parser') #수프 만들기
    for tag in soup.select('div.C4VMK > span > a') :
        if '#' in tag.text :
            hastage_list.append(tag.text)
    driver.find_element_by_link_text('다음').click()
    time.sleep(1)
```

```python
len(hastage_list), hastage_list
```
