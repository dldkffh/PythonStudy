## 한솥메뉴 수집 -> 도시락 이름, 가격 

1. 더보기 버튼 3번 클릭 
2. 모든 메뉴가 나온 페이지를 bs로 변환
3. 수집

```python
!pip install tqdm
```

```python
import time
from tqdm import tqdm_notebook as tn
```

```python
hs_url = 'https://www.hsd.co.kr/menu/menu_list'
driver = webdriver.Chrome()
driver.get(hs_url)
```

```python
for i in range(3):
    driver.find_element_by_class_name('a u_cbox_btn_more').click()
    time.sleep(1)
```

```python
soup = bs(driver.page_source, 'html.parser')
```

```python
menu_name = soup.select('h4.h.fz_03') #h fz_03
menu_price = soup.select('strong') #div strong
```

```python
for i in tn(range(len(menu_name))) :
    print(menu_name[i].text)
    print(menu_price[i].text)
```

# 스타벅스 매정정보 수집
### 매장이름, 주소, 전화번호

1. 서울 전체 지역 페이지 띄우기
2. bs로 현재 페이지 변환
3. 수집

```python
sb_ko_url = 'https://www.istarbucks.co.kr/store/store_map.do'
driver = webdriver.Chrome()
driver.get(sb_ko_url)
```

```python
driver.find_element_by_class_name('loca_search').click()
driver.find_element_by_class_name('set_sido_cd_btn').click()
```

```python
driver.find_element_by_class_name('set_gugun_cd_btn').click()
```

```python
soup = bs(driver.page_source, 'html.parser')
```

```python
sb_name = soup.select('ul.quickSearchResultBoxSidoGugun li strong')
sb_adress = soup.select('ul.quickSearchResultBoxSidoGugun p')
sb_tel = soup.select('ul.quickSearchResultBoxSidoGugun a')
```

```python
sb_names = []
sb_adresses = [] 
sb_tels = []

for i in range(len(sb_adress)) :
    print(sb_name[i].text)
    sb_names.append(sb_name[i].text)
    
    print(sb_adress[i].text[:-len(sb_tel[i].text)])
    #print(sb_adress[i].text.replace(sb_tel[i].text,''))
    sb_adresses.append(sb_adress[i].text[:-len(sb_tel[i].text)])
    
    print(sb_tel[i].text) #sb_adress_s[2] , sb_tel[i].text
    sb_tels.append(sb_tel[i].text)
    
    print('-'*80)
```

```python
len(sb_names), len(sb_adresses), len(sb_tels)
```

```python
import pandas as pd
```

```python
starbuckInfo = {'지점명':sb_names,'주소':sb_adresses,'전화번호':sb_tels}
starbuckDf = pd.DataFrame(starbuckInfo)
starbuckDf.head()
```

```python
starbuckDf.to_csv('starbuck_info.csv', encoding='euc-kr')
```
