```python
money = int(input('금액을 입력하세요.'))
```

```python
print('메뉴를 고르세요.')
menu = int(input('[1]콜라(800원) [2]생수(500원) [3]비타민워터(1500원) >> '))

if menu == 1 :
    price = 800
elif menu == 2 :
    price = 500
elif menu == 3 :
    price = 1500
else :
    print('해당메뉴는 없습니다.')

if money < price :
    print('돈이 부족합니다.')
else :
    money -= price
    
if money >= 1000 :
    ew = money // 1000
else :
    ew = 0
    
print('잔돈 : %d'%(money))
print('천원 : %d'%(ew))
```

```python
num = 0
total = 0 #합계
while num < 10 :
    total += num;
    num = int(input('정수입력 : '))
print('합계 : %d'%total)
```

```python
weight = int(input('현재 몸무게 : '))
wanted = int(input('목표 몸무게 : '))
week = 1

while weight > wanted :
    lost = int(input('%d주차 감량 몸무게 : '%week));
    week += 1;
    weight -= lost

print('%dkg 달성!! 축하합니다.'%weight)
```
